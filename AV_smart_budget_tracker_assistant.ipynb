#Budget Tracking System

## Comparing Spending Against a Budget

In this section, I am creating a budget tracking system that compares the user's total spending against a monthly budget.

This is an important feature because it helps users understand whether they are staying within their planned spending limit or overspending.

The budget tracker will:
- Accept a monthly budget amount
- Compare it with total spending
- Calculate the remaining budget
- Display a clear message for the user

```python
# I am setting a sample monthly budget so I can compare it against the total spending.
# Later, this value can be entered by the user through the Gradio interface.

monthly_budget = 1200

# I am calculating the remaining budget by subtracting total spending from the monthly budget.

remaining_budget = monthly_budget - total_spending

print("Monthly Budget: $", round(monthly_budget, 2))
print("Total Spending: $", round(total_spending, 2))
print("Remaining Budget: $", round(remaining_budget, 2))
```

```python
# I am using an if/else statement to give the user a clear budget status message.
# This helps make the output more meaningful than just showing numbers.

if total_spending > monthly_budget:
    overspent_amount = total_spending - monthly_budget
    print("Budget Status: Over Budget")
    print("You have overspent by $", round(overspent_amount, 2))
else:
    print("Budget Status: Within Budget")
    print("You still have $", round(remaining_budget, 2), "left in your budget.")
```

```python
# I am turning the budget comparison logic into a reusable function.
# This will make it easier to connect the budget tracker to the Gradio interface later.

def check_budget_status(total_spending, monthly_budget):
    remaining_budget = monthly_budget - total_spending

    if total_spending > monthly_budget:
        overspent_amount = total_spending - monthly_budget
        return f"Over Budget: You have overspent by ${overspent_amount:.2f}."
    else:
        return f"Within Budget: You still have ${remaining_budget:.2f} left."
```

```python
# I am testing the budget status function with the current dataset and sample monthly budget.

budget_message = check_budget_status(total_spending, monthly_budget)

print(budget_message)
```

## Budget Insight Message

In this step, I am adding a more detailed budget insight message so the user can understand their budget position more clearly.

Instead of only showing whether the user is over or under budget, this function gives a short explanation based on how much of the budget has been used.


```python
# I am creating a budget insight function to give the user a more meaningful explanation
# of their spending position instead of only showing a basic budget status.

def generate_budget_insight(total_spending, monthly_budget):
    budget_used_percentage = (total_spending / monthly_budget) * 100

    if budget_used_percentage > 100:
        overspent_amount = total_spending - monthly_budget
        return (
            f"You have used {budget_used_percentage:.1f}% of your budget. "
            f"This means you are over budget by ${overspent_amount:.2f}. "
            "You may need to review high-spending categories and reduce non-essential expenses."
        )

    elif budget_used_percentage >= 80:
        remaining_budget = monthly_budget - total_spending
        return (
            f"You have used {budget_used_percentage:.1f}% of your budget. "
            f"You still have ${remaining_budget:.2f} remaining, but you are close to your limit. "
            "It may be a good idea to monitor spending carefully for the rest of the month."
        )

    else:
        remaining_budget = monthly_budget - total_spending
        return (
            f"You have used {budget_used_percentage:.1f}% of your budget. "
            f"You still have ${remaining_budget:.2f} remaining. "
            "You are currently managing your budget well."
        )
```

```python
# I am testing the budget insight function using the current total spending
# and the sample monthly budget to confirm that it gives a clear user-friendly message.

budget_insight = generate_budget_insight(total_spending, monthly_budget)

print(budget_insight)
```

# Savings Goal Calculator

## Creating a Savings Goal Calculator

In this section, I am creating a savings goal calculator that helps users estimate how long it may take to reach a financial target.

This feature is useful because it allows users to:
- Set realistic savings goals
- Understand how regular contributions impact progress
- Receive financial feedback and motivation

The calculator will:
- Accept a savings goal amount
- Accept a monthly savings contribution
- Estimate the number of months required to reach the goal

```python
# I am creating a function that estimates how long it will take
# for a user to reach their savings goal based on monthly contributions.

def calculate_savings_goal(goal_amount, monthly_saving):

    months_required = goal_amount / monthly_saving

    return months_required
```

```python
# I am testing the savings goal calculator using sample values
# to confirm that the function works correctly.

goal_amount = 5000
monthly_saving = 400

months_required = calculate_savings_goal(goal_amount, monthly_saving)

print("Savings Goal: $", goal_amount)
print("Monthly Saving: $", monthly_saving)

print()

print("Estimated Time Required:", round(months_required, 1), "months")
```

## Providing Financial Feedback

In addition to calculating the estimated savings time, I also want the application to provide encouraging financial feedback based on the user's progress plan.

This makes the application feel more interactive and supportive rather than only displaying calculations.

```python
# I am creating a function that gives the user a personalised savings message
# based on how long their savings goal may take to achieve.

def generate_savings_feedback(months_required):

    if months_required <= 6:
        return "Excellent progress. Your savings goal is achievable within a short period of time."

    elif months_required <= 12:
        return "Good progress. Your savings goal is realistic if you continue saving consistently."

    else:
        return "Your savings goal may take longer to achieve. Increasing your monthly savings could help you reach your target faster."
```

```python
# I am testing the financial feedback function
# using the calculated savings timeframe.

feedback_message = generate_savings_feedback(months_required)

print(feedback_message)
```

## Improving the Savings Calculator with Validation

In this step, I am adding simple validation checks to the savings calculator.

This is important because users may enter invalid values, such as a savings goal of zero or a monthly saving amount of zero. These checks help prevent errors and make the tool more reliable.

```python
# I am improving the savings goal calculator by adding basic validation.
# This helps prevent calculation errors if the user enters unrealistic or invalid values.

def calculate_savings_goal_safe(goal_amount, monthly_saving):

    if goal_amount <= 0:
        return "Please enter a savings goal greater than $0."

    elif monthly_saving <= 0:
        return "Please enter a monthly saving amount greater than $0."

    else:
        months_required = goal_amount / monthly_saving
        return f"It will take approximately {months_required:.1f} months to reach your savings goal."
```

```python
# I am testing the improved savings calculator with valid sample values
# to confirm that it produces a useful result.

print(calculate_savings_goal_safe(5000, 400))
```

```python
# I am testing the improved savings calculator with invalid values
# to confirm that the validation messages work correctly.

print(calculate_savings_goal_safe(5000, 0))
print(calculate_savings_goal_safe(0, 400))
```

#Finance Chatbot

## Creating a Basic Finance Chatbot

In this section, I am creating a simple finance chatbot that can respond to basic budgeting questions.

At this stage, the chatbot uses simple keyword matching. This is suitable for an introductory programming project because it shows clear logic using conditional statements.

The chatbot will later be improved and connected to the Gradio interface.

```python
# I am creating a simple finance chatbot function using keyword matching.
# This allows the assistant to respond to common budgeting questions in a beginner-friendly way.

def finance_chatbot(user_question):

    user_question = user_question.lower()

    if "budget" in user_question:
        return "A good budgeting habit is to track your regular expenses and compare them against your monthly income."

    elif "save" in user_question or "saving" in user_question:
        return "A useful savings strategy is to set a realistic goal and save a fixed amount each month."

    elif "food" in user_question:
        return "Food spending can add up quickly, so reviewing grocery, takeaway, and coffee expenses may help reduce costs."

    elif "transport" in user_question:
        return "Transport costs can be managed by comparing fuel, public transport, parking, and ride-share expenses."

    elif "entertainment" in user_question:
        return "Entertainment spending is important for lifestyle balance, but it should be monitored so it does not exceed the budget."

    else:
        return "I can help with budgeting, saving, spending categories, and basic financial planning questions."
```

```python
# I am testing the chatbot with sample finance questions
# to check whether the keyword responses work correctly.

print(finance_chatbot("How can I improve my budget?"))
print(finance_chatbot("How can I save more money?"))
print(finance_chatbot("Am I spending too much on food?"))
```

## Making the Chatbot Use Transaction Data

In this step, I am improving the chatbot so it can respond using information from the transaction dataset.

This helps make the assistant more useful because it can answer simple questions about the user's actual spending patterns instead of only giving general budgeting advice.

```python
# I am creating an improved chatbot function that can use the transaction data.
# This allows the chatbot to answer some questions using actual spending calculations.

def finance_chatbot_with_data(user_question):

    user_question = user_question.lower()

    if "total" in user_question or "spent" in user_question:
        return f"Your total spending is ${total_spending:.2f}."

    elif "highest" in user_question or "most" in user_question:
        return f"Your highest spending category is {highest_category}, with ${highest_amount:.2f} spent."

    elif "average" in user_question:
        return f"Your average transaction amount is ${average_transaction:.2f}."

    elif "budget" in user_question:
        return generate_budget_insight(total_spending, monthly_budget)

    elif "save" in user_question or "saving" in user_question:
        return "You can use the savings goal calculator to estimate how long it will take to reach a financial target."

    else:
        return finance_chatbot(user_question)
```

```python
# I am testing the improved chatbot using questions that relate to the dataset.
# This helps confirm that the chatbot can use calculated spending information.

print(finance_chatbot_with_data("How much have I spent in total?"))
print(finance_chatbot_with_data("What category did I spend the most on?"))
print(finance_chatbot_with_data("What is my average transaction amount?"))
```

# Retrieval-Augmented Generation (RAG)

## Retrieving Information from Transaction Data

In this section, I am adding a simple retrieval system that searches the transaction dataset and returns relevant information based on the user's question.

This is a beginner-friendly version of RAG because the assistant retrieves information from the uploaded data before generating a response.

The goal is to help the assistant answer questions such as:
- How much did I spend on food?
- What did I spend on transport?
- Which category has the highest spending?

```python
# I am creating a simple retrieval function that checks whether the user's question
# mentions one of the spending categories in the transaction dataset.

def retrieve_category_spending(user_question):

    user_question = user_question.lower()

    for category in df["Category"].unique():

        if category.lower() in user_question:
            category_total = df[df["Category"] == category]["Amount"].sum()

            return f"You spent ${category_total:.2f} on {category}."

    return "I could not find a matching spending category in your question."
```

```python
# I am testing the retrieval function with sample questions
# to confirm that it can find spending information from the dataset.

print(retrieve_category_spending("How much did I spend on food?"))
print(retrieve_category_spending("What did I spend on transport?"))
print(retrieve_category_spending("How much did I spend on shopping?"))
```

## Combining Retrieval with Chatbot Responses

In this step, I am combining the retrieval function with the chatbot so the assistant can answer questions using both general finance logic and transaction-specific information.

```python
# I am creating a RAG-style chatbot function.
# It first tries to retrieve relevant transaction information from the dataset.
# If no specific category is found, it falls back to the finance chatbot response.

def rag_finance_assistant(user_question):

    retrieved_answer = retrieve_category_spending(user_question)

    if retrieved_answer != "I could not find a matching spending category in your question.":
        return retrieved_answer

    else:
        return finance_chatbot_with_data(user_question)
```

```python
# I am testing the RAG-style finance assistant with both category-specific
# and general finance questions.

print(rag_finance_assistant("How much did I spend on food?"))
print(rag_finance_assistant("What category did I spend the most on?"))
print(rag_finance_assistant("How can I improve my budget?"))
```

# Gradio User Interface

## Building the Interactive Interface

In this section, I am creating a Gradio interface so users can interact with the Smart Budget Tracker Assistant more easily.

The interface will allow users to:
- Ask finance-related questions
- Receive responses from the assistant
- Use the RAG-style transaction retrieval function

This turns the project from a notebook-based analysis into a simple interactive application.

```python
# I am creating a function that connects the user's question
# to the RAG finance assistant built earlier in the notebook.

def gradio_chatbot_response(user_question):
    return rag_finance_assistant(user_question)
```

This allows the user to enter a savings goal and monthly saving amount directly into the app and receive a result.

```python
# I am creating a Gradio-friendly savings calculator function.
# This connects the savings goal tool to the user interface.

def gradio_savings_calculator(goal_amount, monthly_saving):
    return calculate_savings_goal_safe(goal_amount, monthly_saving)
```
