#Data Cleaning and Validation

## Validating and Cleaning the Dataset

Before beginning any financial analysis, I need to validate and clean the dataset to ensure the information is accurate and usable.

This step is important because:
- Missing values can cause calculation errors
- Incorrect data types may break analysis functions
- Invalid transaction amounts can produce inaccurate results

By validating the dataset early, I can reduce errors later in the project and improve the reliability of the application.

```python
# I want to check whether the dataset contains any missing values
# before starting the spending analysis.

print("Missing Values in Each Column:")
print()

print(df.isnull().sum())
```

```python
# I am checking the data types of each column
# to ensure the dataset has loaded correctly.

print("Dataset Data Types:")
print()

print(df.dtypes)
```

```python
# I want to make sure the Amount column is stored as a numeric value
# so mathematical calculations can be performed correctly.

df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

print("Amount column successfully converted to numeric format.")
```

```python
# I am checking whether any invalid values were created
# during the numeric conversion process.

invalid_amounts = df["Amount"].isnull().sum()

print("Number of invalid amount values:", invalid_amounts)
```

```python
# I want to display a cleaned preview of the dataset
# before moving into the spending analysis stage.

df.head()
```

#Spending Analysis Functions

## Analysing Spending Behaviour

In this section, I am creating functions to analyse the user's spending behaviour using the uploaded transaction dataset.

The goal of this analysis is to:
- Calculate total spending
- Identify the highest spending categories
- Understand where money is being spent most frequently
- Generate useful financial insights

These functions will later support:
- Budget tracking
- Savings recommendations
- Chatbot responses
- Data visualisations

```python
# I am calculating the user's total spending
# by summing all transaction amounts in the dataset.

total_spending = df["Amount"].sum()

print("Total Spending: $", round(total_spending, 2))
```

```python
# I want to analyse how much money was spent in each category.
# Grouping the data by category helps identify spending patterns.

category_spending = df.groupby("Category")["Amount"].sum()

print("Spending by Category:")
print()

print(category_spending)
```

```python
# I am identifying the category with the highest spending
# to better understand the user's largest expense area.

highest_category = category_spending.idxmax()

highest_amount = category_spending.max()

print("Highest Spending Category:", highest_category)
print("Amount Spent: $", round(highest_amount, 2))
```

```python
# I want to calculate the average transaction amount
# to better understand typical spending behaviour.

average_transaction = df["Amount"].mean()

print("Average Transaction Amount: $", round(average_transaction, 2))
```

```python
# I am identifying the largest individual transaction
# to highlight unusually high expenses in the dataset.

largest_transaction = df.loc[df["Amount"].idxmax()]

print("Largest Transaction:")
print()

print(largest_transaction)
```
